```python
df=pd.read_csv('dataset 1.csv')
df2=pd.read_excel('dataset 2.xlsx')
df.head()
```

```python
df2.head()
```

```python
merge_df=pd.merge(df,df2,on='Policy no.',how='inner')
```

```python
merge_df.groupby('sex').mean()
```

```python
#find avg age of people in every region
#find median bmi for male and female records
#find min and max age,median bmi,mode of children,count of smoker for male and female records
#read about groupby on multiple columns
```

```python
merge_df.groupby('region').mean()
```

```python
merge_df.groupby('region')['age'].mean()
```

```python
merge_df.groupby('sex')['age'].min()
```

```python
merge_df.groupby('sex').agg(['min','max','mean','median'])
```

```python
merge_df.groupby('sex')['bmi'].median()
```

```python
merge_df['children'].mode()  
```

```python
merge_df.groupby('sex')['smoker'].count()
```

```python
merge_df.groupby('sex').agg({'age':['min','max'],'bmi':'median','smoker':'count','children':'mean'})
```

```python
merge_df['children'].mode()[0]
```

## Performing encoding on smoker column

```python
merge_df['smoker']=merge_df['smoker'].map({'yes':1,'no':0})
```

```python
merge_df.head()
```

```python
merge_df.groupby('sex').agg({'age':['min','max'],'bmi':'median','smoker':'sum','children':'mean'})
```
