#  Concat,Merge and Join operations on dataframe
    -Concat is used to append one or more dataframes below the other dataframes(or sideways,depending on whether the axis option is set to 0 or 1)
    -Merge: Pandas has full featured,high performance in memory join operations which is very similar to databases.
    -Join is a convinient method for combining the columns of two potentially different indexed data frames into a single result data frame

##  Concatination with examples

```python
a = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    },
    index=[0, 1, 2, 3],
)
```

```python
a
```

```python
b = pd.DataFrame(
    {
        "A": ["A4", "A5", "A6", "A7"],
        "B": ["B4", "B5", "B6", "B7"],
        "C": ["C4", "C5", "C6", "C7"],
        "D": ["D4", "D5", "D6", "D7"],
    },
    index=[4, 5, 6, 7],
)
```

```python
b
```

```python
pd.concat([a,b])
```

```python
pd.concat([a,b],axis=1)
```

##  Join with examples

```python
left = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2"],
        "B": ["B0", "B1", "B2"] },
    index=['k0', 'k1', 'k2']
)
left
```

```python
right = pd.DataFrame(
    {
        "C": ["C0", "C1", "C2"],
        "D": ["D0", "D1", "D2"] },
    index=['k0', 'k2', 'k3']
)
right
```

```python
result=left.join(right)
```

```python
result
```

```python
right_result=right.join(left)
right_result
```

```python
inner_result=left.join(right,how='inner')
inner_result
```

```python
outer_result=left.join(right,how='outer')
outer_result
```
