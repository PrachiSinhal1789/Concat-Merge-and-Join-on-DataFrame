## Loc
    -Loc function can be used to access columns by column name rather than indexes and it is inclusive of the end index
    -we cant use reverse indexing in Loc

```python
df.loc[0:4,:]   #like head
```

```python
df.loc[0:1,'age':'bmi']
```

```python
df.loc[0:3,'charges in INR']
```

```python
df.loc[1334:1338,['Policy no.','age','bmi','charges in INR']] #IGNORE SEx column
```

```python
df.loc[1,['Policy no.','age']]
```
