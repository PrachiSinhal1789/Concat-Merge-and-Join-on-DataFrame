#### HISTOGRAM
     Histograms are used to analyse one variable or a feature at a time.We use a histogram when we want to see the    count of a variable on a chart.

```python
plt.hist(df['No. of Clicks'])
plt.title('Histogram of No. of Clicks')
```

```python
plt.hist(df['No. of Clicks'])
plt.title('Histogram of No. of Clicks')
plt.xlabel('No. of Clicks')
plt.ylabel('Count of Clicks')
plt.show()
```

```python
sns.displot(df['No. of Clicks'])
```

```python
sns.distplot(df['No. of Clicks']) #this is not binned
```

```python
sns.distplot(df['No. of Clicks'],bins=10)
```

```python
df['No. of Clicks'].plot(kind='hist') #using pandas for plotting
```

```python
df['Pages Viewed'].plot(kind='hist',bins=7)
pl.xlabel('Pages Viewed')
pl.title('Histogram of Pages Viewed')
```

```python
plt.hist(df['Time (min)'])
plt.xlabel('Time (min)')
plt.ylabel('Sum of Time(min)')
plt.title('Histogram of Time(min)')
plt.show()
```

```python
sns.distplot(df['Amount Spent (Rs.)']).set(title='Histogram of Amount Spent (Rs.)')
```

```python
sns.histplot(df['Pages Viewed']).set(title='Histogram of Pages Viewed')
```

##  Pie Chart
    -Pie Chart is a circular chart which is divided into slices and is used to view and compare categorical columns.

```python
df['Browser'].unique()
```

```python
df['Browser'].value_counts()
```

```python
plt.pie(df['Browser'].value_counts())
```

```python
plt.pie(df['Browser'].value_counts(),labels=['Chrome','Firefox','IE'])
```

```python
plt.pie(df['Browser'].value_counts(),labels=['Chrome','Firefox','IE'])
plt.show()#to hide text
```

```python
plt.pie(df['Browser'].value_counts(),labels=['Chrome','Firefox','IE'],autopct='%.2f')
plt.show()
```

```python
plt.pie(df['Browser'].value_counts(),labels=['Chrome','Firefox','IE'],autopct='%.2f%%')
plt.show()
```

```python
df['Browser'].value_counts().plot(kind='pie',autopct='%.2f%%')
```

```python
colors = sns.color_palette('dark')  #you can use bright
plt.pie(df['Browser'].value_counts(),colors=colors,labels=['Chrome','Firefox','IE'],autopct='%.2f%%')
plt.show()
```

##  Bar plot/graph
    -A Bar plot is used to analyse several groups of data in a column.

```python
Day_valuecounts=df['Day'].value_counts()
```

```python
df['Day'].value_counts().index
```

```python
Day_valuecounts.index
```

```python
Day_valuecounts.values
```

```python
plt.bar(Day_valuecounts.index,Day_valuecounts.values)
```

```python
plt.figure(figsize=(7,10)) #chk which is height and width
plt.bar(Day_valuecounts.index,Day_valuecounts.values)
plt.xlabel('Days of the week')
plt.ylabel('Count')
plt.title('Bar graph')
```

```python
sns.barplot(x=df['Day'].value_counts().index,y=df['Day'].value_counts().values).set(title='Bar graph',xlabel='Days',ylabel='Count of Days')
```

```python
df['Day'].value_counts().plot(kind='bar',figsize=(4,6))
pl.xlabel('Days')
pl.ylabel('Count')
pl.title('Bar graph')
```

```python
df['Day'].value_counts().plot(kind='barh',figsize=(6,4))
```

## Boxplot
    -Boxplot is used commonly to study the descriptive statistics of a column.
    -It is mostly used to plot the outlier in the column.

```python
plt.boxplot(df['Amount Spent (Rs.)']) #lower wisker lw=q1-1.5*iqr uw=q3+1.5*iqr where iqr=q3-q1
plt.show()
#dot above up is outlier
```

```python
df['Amount Spent (Rs.)'].plot(kind='box')
```

```python
plt.boxplot(df['Pages Viewed'])
plt.show()
```

```python
plt.boxplot(df['No. of Clicks'])
plt.show()
```
