### Step 1. Import the necessary libraries

```python
import pandas as pd
import datetime
import io
import requests
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data)

```python
url='https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data'
s=requests.get(url).content
df=pd.read_csv(io.StringIO(s.decode('utf-8')),delim_whitespace=True)
```

### Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index.

```python
data=df.copy()
data['Date']=df['Dy'].astype('str')+'-'+df['Mo'].astype('str')+"-"+df['Yr'].astype('str')
data['Date']=pd.to_datetime(data['Date'])
data.drop(['Yr','Mo','Dy'],axis=1,inplace=True)
data
```

### Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python
def fix_date(x):
    if x.year > 1979:
        year = x.year - 100
    else:
        year = x.year
    return datetime.date(year,x.month,x.day)
data['Date'] = data['Date'].apply(fix_date)
data
```

### Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days 

#### A different set of numbers for each location.

```python
min_value=data.min()
max_value=data.max()
mean_value=data.mean()
std_value=data.std()
loc_stats=pd.concat([min_value,max_value,mean_value,std_value],axis=1)
loc_stats.rename(columns={0:'min',1:'max',2:'mean',3:'std'})
```

### Step 10. Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day.

#### A different set of numbers for each day.

```python
data1=data.transpose()
data1
```

```python
min_value=data1.min()
max_value=data1.max()
mean_value=data1.mean()
std_value=data1.std()
day_stats=pd.concat([min_value,max_value,mean_value,std_value],axis=1)
day_stats.rename(columns={0:'min',1:'max',2:'mean',3:'std'})
```

### Step 11. Find the average windspeed in January for each location.  
#### Treat January 1961 and January 1962 both as January.

```python
Jan1 = data.loc['1961-01']
Jan2 =  data.loc['1962-01']
final = pd.concat([Jan1,Jan2])
for i in final.columns:
    print(f'Average windspeed for "{i}" for Jan-61 and Jan-62 is: {round(final[i].mean(),2)}')
```

### Step 15. Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks.

```python
filter_days = data.loc['1961-01-02':'1962-01-01']   #there are 52 weeks in a year
weekly_min = filter_days.resample('W').min()
weekly_max = filter_days.resample('W').max()
weekly_mean = round(filter_days.resample('W').mean())
weekly_std = round(filter_days.resample('W').std())
print(f'Weekly Min\n{weekly_min.min(axis = 1)}')
print(f'\nWeekly Max\n{weekly_max.max(axis = 1)}')
print(f'\nWeekly Mean\n{weekly_mean.mean(axis = 1)}')
print(f'\nWeekly Std\n{weekly_std.std(axis = 1)}')
```
